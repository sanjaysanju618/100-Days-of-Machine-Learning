## [Matplotlib Graphing Intro](https://youtu.be/H72jSxkLQHQ)

```python
plt.plot([5, 6, 7, 8], [7, 2, 3, 8])
plt.show()
```

## [Matplotlib Labels and titles](https://youtu.be/B06l-lGU-0Q)

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
```

```python
plt.plot(x, y)
plt.title("Sample Chart")
plt.xlabel("Natural number")
plt.ylabel("Multiples number")
plt.show()
```

## [Matplotlib Styles](https://youtu.be/sCaGYsEYy-k)

```python
x1 = [1, 2, 3, 4]
y1 = [3, 6, 9, 12]
```

```python
plt.plot(x, y, 'g', linewidth=5)
plt.plot(x1, y1)
plt.title("Sample Chart")
plt.xlabel("Natural number")
plt.ylabel("Multiples number")
plt.show()
```

## [Matplotlib legends](https://youtu.be/X4_LraGY_VQ)

```python
plt.plot(x, y, 'g', linewidth=5, label='2 Multiples')
plt.plot(x1, y1, label='3 Multiples')
plt.title("Sample Chart")
plt.xlabel("Natural number")
plt.ylabel("Multiples number")
plt.legend() # reflect....
plt.show()
```

```python
plt.plot(x, y, 'g', linewidth=5, label='2 Multiples')
plt.plot(x1, y1, label='3 Multiples')
plt.title("Sample Chart")
plt.xlabel("Natural number")
plt.ylabel("Multiples number")
plt.legend() # reflect....

plt.grid(True, color='red') # show grid....
plt.show()
```

## [Scatter plots and bar charts](https://youtu.be/MAMu8wawNo8)

```python
plt.plot(x, y, 'g', linewidth=5, label='2 Multiples')
plt.scatter(x1, y1, color='blue' ,label='3 Multiples') # scatter....
plt.title("Sample Chart")
plt.xlabel("Natural number")
plt.ylabel("Multiples number")
plt.legend() 
plt.grid(True, color='red')
plt.show()
```

```python
plt.plot(x, y, 'g', linewidth=5, label='2 Multiples')
plt.bar(x1, y1, color='blue' ,label='3 Multiples') # bar....
plt.title("Sample Chart")
plt.xlabel("Natural number")
plt.ylabel("Multiples number")
plt.legend() 
plt.grid(True, color='red')
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [2, 4, 6, 8]
x1 = [3, 5, 7, 9]
y1 = [3, 6, 9, 12]
```

```python
plt.bar(x, y, color='green',label='2 Multiples', align='center')
plt.bar(x1, y1, color='blue' ,label='3 Multiples', align='center') # bar....
plt.title("Sample Chart")
plt.xlabel("Natural number")
plt.ylabel("Multiples number")
plt.legend() 
plt.grid(True, color='red')
plt.show()
```
